### Điểm đánh giá

```python
data['Điểm đánh giá'] = data['Điểm đánh giá'].str.replace(',','.')
data['Điểm đánh giá'] = data['Điểm đánh giá'].str.strip()
data['Điểm đánh giá'] = data['Điểm đánh giá'].astype('float')
data
```

```python
for i in range(len(data)):
    a=data['Quốc gia'].iloc[i]
    if a.find('từ ')>1:
        data['Quốc gia'].iloc[i]=a[a.find('từ ')+3:]
    else:
        data['Quốc gia'].iloc[i]=np.nan
```

```python
for i in range(len(data)):
    a=data['Thời gian'].iloc[i]
    if a.find('ở ')>1:
        data['Thời gian'].iloc[i]=a[a.find('ở ')+2:a.find('ở ')+8].strip()
    else:
        data['Thời gian'].iloc[i]=np.nan
```
